```python
import json

import pandas as pd

raw_cols = "题型(选填)	题目/题干/小题题干	选项A	选项B	选项C	选项D	选项E	选项F	选项G	选项H	选项I	选项J	答案	解析	章节	知识点	分值"
questions_df = pd.DataFrame(columns=raw_cols.split("\t"))
```

```python
def add_question(question_type=None,
                 question_text="",
                 answer_a=None,
                 answer_b=None,
                 answer_c=None,
                 answer_d=None,
                 answer_e=None,
                 answer_f=None,
                 answer_g=None,
                 answer_h=None,
                 answer_i=None,
                 answer_j=None,
                 key_text="",
                 analysis_text="",
                 chapter_text="",
                 knowledge_text="",
                 score=1):
    new_row = {
        "题型(选填)": question_type,
        "题目/题干/小题题干": question_text,
        "选项A": answer_a,
        "选项B": answer_b,
        "选项C": answer_c,
        "选项D": answer_d,
        "选项E": answer_e,
        "选项F": answer_f,
        "选项G": answer_g,
        "选项H": answer_h,
        "选项I": answer_i,
        "选项J": answer_j,
        "答案": key_text,
        "解析": analysis_text,
        "章节": chapter_text,
        "知识点": knowledge_text,
        "分值": score
    }
    questions_df.loc[len(questions_df)] = new_row
```

```python
keys_list = "B C A B B B C D D B B A B A D C B D B C C C B D C A D B A D C A B B A".split()
```

```python
questions_df = pd.DataFrame(columns=raw_cols.split("\t"))

question_file = open("./英语题目.txt", "r", encoding="utf-8")

line = question_file.readline()
number = 1

while line:

    if line and line.startswith(f"{number}."):  # 题目字段

        question_text = line[line.find(".") + 1:].strip()  # 这个是题目，下一行是选项
        answers_list = [_[_.find(".") + 1:].strip() for _ in question_file.readline().split("\t") if len(_) > 0]
        add_question(question_type="单选题",
                     question_text=question_text,
                     answer_a=answers_list[0],
                     answer_b=answers_list[1],
                     answer_c=answers_list[2],
                     answer_d=answers_list[3],
                     key_text=keys_list[number - 1])

        number += 1

    line = question_file.readline()
```

```python
with open("tiku.json", "r", encoding="utf-8") as f:
    data = json.load(f)

    for question in data:
        answers_list = question["options"]

        add_question(question_type="单选题",
                     question_text=question.get("title"),
                     answer_a=answers_list[0],
                     answer_b=answers_list[1],
                     answer_c=answers_list[2],
                     answer_d=answers_list[3],
                     key_text=question['key'],
                     analysis_text=question['tip'])
```

```python
questions_df.to_excel("导出题库.xlsx", index=False)
```
